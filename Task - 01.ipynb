#### 4. Remove Outlier

```python
Q1 = df['SibSp'].quantile(0.25)
Q3 = df['SibSp'].quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
# Assuming both lower_bound and upper_bound are defined
s_cln = df[(df['SibSp'] >= lower_bound) & (df['SibSp'] <= upper_bound)]
```

```python
sns.boxplot(y='SibSp', data=s_cln)
```

```python
Q1 = df['Parch'].quantile(0.25)
Q3 = df['Parch'].quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
# Assuming both lower_bound and upper_bound are defined
s_cln = df[(df['Parch'] >= lower_bound) & (df['Parch'] <= upper_bound)]
```

```python
sns.boxplot(y='Parch', data=s_cln)
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
# Assuming both lower_bound and upper_bound are defined
s_cln = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]
```

```python
sns.boxplot(y='Fare', data=s_cln)
```

```python
age2 = s_cln.groupby('Age').mean(numeric_only=True)
```

```python
age2['SibSp'].plot(kind='line')

plt.title('Mean of SibSp by Age')
plt.xlabel('Age')
plt.ylabel('Mean of SibSp')
plt.show()
```

```python
age2['Fare'].plot(kind='line')

plt.title('Mean of Fare by Age')
plt.xlabel('Age')
plt.ylabel('Mean of Fare')
plt.show()
```

### Analyze Relationships

#### a. Survival Rate by Gender

```python
# Survival count by gender
sns.countplot(x='Survived', hue='Sex', data=s_cln)
plt.title('Survival Count by Gender')
plt.show()
```

#### b. Survival Rate by Passenger Class

```python
# Survival count by passenger class
sns.countplot(x='Pclass', hue='Survived', data=s_cln)
plt.title('Survival Count by Passenger Class')
plt.show()
```

#### c. Age Distribution by Survival Status

```python
# Age distribution for survived vs not survived
plt.figure(figsize=(10, 5))
sns.histplot(data=s_cln, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Age Distribution by Survival Status')
plt.show()
```

#### d. Fare Distribution

```python
# Fare distribution by survival status
plt.figure(figsize=(10, 5))
sns.barplot(x='Survived', y='Fare', data=s_cln)
plt.title('Fare Distribution by Survival Status')
plt.show()
```

#### e. Correlation Analysis

```python
numeric_columns = df.select_dtypes(include=['number']).columns
corr = s_cln[numeric_columns].corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

### Observations

#### Fare vs Pclass: There is a strong negative correlation (-0.59), indicating that passengers in higher classes (lower Pclass values) tend to have higher fares.
#### SibSp and Parch: There is a moderate positive correlation (0.41), suggesting that passengers with more siblings/spouses (SibSp) also tend to have more parents/children onboard (Parch).
#### Fare vs Survived: There is a mild positive correlation (0.23), indicating that passengers who paid higher fares had a slightly better chance of survival.
#### Pclass vs Survived: A negative correlation (-0.24), which aligns with the idea that passengers in higher classes (lower Pclass values) had a better survival rate.
